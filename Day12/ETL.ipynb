```python
import pandas as pd
import sqlalchemy as ssl
```

```python
engine = ssl.create_engine('postgresql+psycopg2://postgres:root\
@localhost/postgres')
conn= engine.connect()
```

```python
def extract():
    df_orders = pd.read_csv('/Users/B0264653/Desktop/Namaste-Python/Day8/orders.txt')
    df_returns = pd.read_csv('/Users/B0264653/Desktop/Namaste-Python/Day10/returns.txt')
    return df_orders , df_returns  

def transform(df_orders,df_returns ):
    df = pd.merge(df_orders, df_returns, how='inner', left_on='order_id', right_on ='order_id')
    return df

def load(df):
    df.to_sql('etl1',con=conn, index=False, if_exists='append') #append/replace
    conn.commit() #commit and freeze if not connect then locks the table
```

```python
#extract
df_orders , df_returns  = extract()

#transform
df = transform(df_orders , df_returns)

#load
load(df)
```
