```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/B0264653/Desktop/Namaste-Python/Day8/orders.txt')
```

```python
df
```

```python
# 1- create a new column in the orders dataframe 
# with name profit_flag. Set its value as 1 if profit 
# is greater than 0 else set 0.

df['profit_flag']=''
df.loc[(df['profit'] > 0) , 'profit_flag'] = 1
df.loc[(df['profit'] < 0) , 'profit_flag'] = 0
df
```

```python
# 2- in the orders data frame rename column names to
# upper case ..eg order_id should be ORDER_ID

df.columns
df.rename(columns={'order_id':'ORDER_ID', 'order_date':'ORDER_DATE', 'customer_name':'CUSTOMER_NAME', 'city':'CITY', 'category':'CATEGORY','product_id':'PRODUCT_ID', 'sales':'SALES', 'profit':'PROFIT', 'profit_flag':'PROFIT_FLAG'})
```

```python
# 3- again change the column names back to lower case
df
```

```python
# 4- sort the dataframe column by profit in descending order 

df.sort_index(ascending=False)
```

```python
df.columns
```

```python
# 5- sort the dataframe column by sales in ascending order , 
# in case of a tie it should be sorted by profit in descending order 

df.sort_values(by=['order_id','profit'], ascending=[True,False])
```

```python
# 6- reset the index to original state

df.reset_index()
```

```python
# 7- create a column as profit_category . sets it value as below:

# if profit < 0 then 'loss'
# if profit between 0 and 100 then 'low profit' 
# if profit between 100 and 500 then 'medium profit'
# else high profit 


df['profit_category']=''
df.loc[(df['profit'] < 0) , 'profit_category'] = 'loss'
df.loc[(df['profit'] > 0) & (df['profit'] < 100) , 'profit_category'] = 'low profit'
df.loc[(df['profit'] > 100) & (df['profit'] < 500) , 'profit_category'] = 'medium profit'
df
```

```python
# 8- create a new column sales_profit_ratio and set its value as (profit/sales)*100

df['sales_profit_ratio']= (df['profit']/df['sales'])*100
df
```

```python
# 9- find all the orders where customer name starts with 'D' and ends with 'e'

df9 = (df['customer_name'].str.startswith('D')) & (df['customer_name'].str.endswith('e')) 
df[df9]
```

```python
# 10- create a new column order_year and populate it with year of order_date. 
# ( hint : use the series objects slice method )
 
OD = df['order_date'].str.slice(0,4)
df['order_year'] = OD
df
```

```python
# 11 - delete column sales_profit_ratio from the dataframe

df.drop(columns='sales_profit_ratio')
```

```python
# 12- delete all the rows where profit is negative 

m = df['profit'] < 0
df[m]

df.drop(index = df[m].index,inplace=True )
df
```
