```python
import pandas as pd
```

```python
###Read the iphone.csv dataset to pandas dataframe
df_iphone = pd.read_csv('/Users/B0264653/Desktop/Namaste-Python/project_2/iphone.csv')
df_iphone
```

```python
df_iphone.columns
```

```python
# 1 - The column names have spaces . rename the column names 
# to have underscore '_' instead of space (try to do in one go instead 
# of specifying each column nam in rename method)

df_iphone.rename (columns={'Product Name':'Product_Name', 'Product URL':'Product_URL','Sale Price': 'Sale_Price','Number Of Ratings': 'Number_Of_Reviews', 'Star Rating':'Star_Rating'},inplace=True)
df_iphone
```

```python
# 2- start rating for some of the models is missing in
#  the dataset. fill those missing values with the average
#   rating all the models.

df_star_rating_mean = df_iphone['Star_Rating'].mean()
print(df_star_rating_mean)
df_iphone_d= df_iphone['Star_Rating'].fillna(value=df_star_rating_mean,inplace=True)
df_iphone
```

```python
# 3- Now instead of filling missing values with avg rating of full dataset ,
# fill with avg rating based on RAM. example :  if rating for a 2 gb phone
# is missing then take average of all other 2 gb phones rating and 
# fill that value. 

df_star_rating_mean_grp = df_iphone.groupby('Ram')['Star_Rating'].mean()
print(df_star_rating_mean_grp)
df_iphone_grp= df_iphone['Star_Rating'].fillna(value=df_star_rating_mean,inplace=True)
df_iphone
```

```python
# 4- create a new column in the dataframe "Discount_Percentage" 
# based on MRP and sale value
df_iphone['Discount_Percentage'] = (df_iphone['Sale_Price']/df_iphone['Mrp'])*100
df_iphone
```

```python
# 5- which model has highest percent discount ?
df_iphone_max= df_iphone.groupby('Product_Name')['Discount_Percentage'].max()
df_iphone_max
```

```python
# 6- find total no of models  each space configuration (128 GB , 64 GB etc)
df_iphone_max= df_iphone.groupby('Ram')['Product_Name'].count()
df_iphone_max
```

```python
# 7- find total number of models for each color 
df_iphone[["model","color1"]] = df_iphone['Product_Name'].str.split("(", expand = True)
df_iphone.groupby('model')['model'].count()
```

```python
df_iphone[["color","specify"]] = df_iphone['color1'].str.split(",", expand = True)
df_iphone.drop(['color1','specify'],axis=1,inplace=True)
```

```python
df_iphone['version'] = df_iphone['model'].str[6:15]
df_iphone
```

```python
# 8- find total number of models by iphone version 

df_iphone.groupby('version')['model'].count()
```

```python
# 9- list top 5 models having highest no of reviews 

df_iphone.groupby('model')['Number_Of_Reviews'].sum().sort_values(ascending=False).head(5)
```

```python
# 10 - what is the price diffrence between highest price and
# lowest price iphone (based on mrp)

df_iphone_max = df_iphone['Mrp'].max()
df_iphone_min = df_iphone['Mrp'].min()
df_iphone_diff = df_iphone_max -  df_iphone_min
df_iphone_diff
```
