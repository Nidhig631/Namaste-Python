```python
# 1- write a program to fetch all the orders with indexes
#   multiple of 100 eg : 100,200,300,400..... . Display order_id ,
# category and sales columns

import pandas as pd
df = pd.read_csv('/Users/B0264653/Desktop/Namaste-Python/Day8/orders.txt')
#get index 
df_filtered = df[df.index%100==0]
col = ['order_id','category','sales']
df_100_multiple = df_filtered[col]

df_100_multiple
```

```python
# 2- write a program to find all orders in furniture category 
# with sales amount between 1000 and 1200

df_furn = (df['category']=='Furniture') & (df['sales'] >= 1000) & (df['sales'] <= 1200)
df[df_furn]
```

```python
# 3- write a program to find all the orders where there was a loss (profit<0) 
# in the city of new york , Los Angeles and Seattle

df_filtered = (df['profit'] < 0) & (df['city'].isin(['New York City','Seattle','Los Angeles']))
df_filtered
df[df_filtered]
```
